```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
def summary(df) :
    tf = pd.DataFrame({'데이터형태(dtypes)' : df.dtypes,
                       '결측치 수(null)' : df.isnull().sum(),
                       '결측치 비율(%)' : (df.isnull().sum()/len(df)).round(2),
                       '고유값 수(nunique)' : df.nunique(),
                      })
    for col in tf.index:
        tf.loc[col, '자료 내용(contents)'] = str(df[col].unique()[0:5])
        tf.loc[col, 'nan 비율'] = int(df[df[col] == 'nan'].shape[0]/len(df)*100)
        
    return tf
```

```python
files = [ '../data\\2016년 상반기.xlsx',
 '../data\\2016년 하반기.xlsx',
 '../data\\2017년 상반기.xlsx',
 '../data\\2017년 하반기.xlsx',
 '../data\\2018년 상반기.xlsx',
 '../data\\2018년 하반기.xlsx',
 '../data\\2019년 상반기.xlsx',
 '../data\\2019년 하반기.xlsx',
 '../data\\2020년 상반기.xlsx',
 '../data\\2020년 하반기.xlsx',
 '../data\\2021년.xlsx']
```

```python
%%time

from tqdm import tqdm_notebook # 파일 합치기 시간이 걸리니 모듈 호출 conda install tqdm

tf = pd.DataFrame()     # 빈 데이터 프레임을 만들어 아래 반복문으로 파일을 하나씩 불러 append하는 방법
sum = 0                 # 최종 데이터셋의 전체 인스턴스 확인용

for file in  tqdm_notebook(files):
    print(file)
    data = pd.read_excel(file) #usecols = mycol) # 필요한 컬럼(mycol)만 읽어 data 객체에 저장
    data = data[(data['검사종류 대분류명'] == '정밀') | (data['검사종류 대분류명'] == '무작위표본')]
    sum += data.shape[0]                        # 11개의 데이터프레임의 크기를 누적하여 총 인스턴스를 확인
    tf = tf.append(data)                        # 11개의 데이터프레임을 반복문으로 append
    print('*'*100)
    
print(f'총 instance는 {sum}개') # check digits
tf.shape
tf.head()
```

```python
tf.to_pickle('../pkl/total-import-food-16-21.pkl')
```
